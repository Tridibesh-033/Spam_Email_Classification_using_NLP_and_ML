```python
ds["spam"]=ds["Category"].map({'ham':0,'spam':1})
```

```python
ds1=ds.drop('Category',axis=1)
ds1.head()
```

```python
ds2=ds1.drop_duplicates(subset=['Message'])
ds2.shape
```

```python
ds2.head()
```

```python
# Convert to lowercase
ds2['clean_text']= ds2['Message'].str.lower()
ds2.head()
```

```python
import string
string.punctuation
```

```python
# Remove of Punctuations function

def remove_punctuation(text):
    punctuations=string.punctuation
    return text.translate(str.maketrans('','',punctuations))
```

```python
# call punctuation function
ds2['clean_text']=ds2['clean_text'].apply(lambda x: remove_punctuation(x))
```

```python
# Stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_list=stopwords.words('english')
stopwords_list
```

```python
# # Stopwords function

import numpy
import gensim
from nltk.corpus import stopwords

# Load English stopwords from NLTK
stopwords_list = set(stopwords.words('english'))

# Define the function to remove stopwords
def remove_stopwords(text):
    # gensim.utils.simple_preprocess produces a list, not a generator
    tokens = gensim.utils.simple_preprocess(str(text))  # Ensure input is a string
    filtered_tokens = [word for word in tokens if word not in stopwords_list]
    return ' '.join(filtered_tokens)  # Join tokens back into a string

```

```python

ds2['clean_text']=ds2['clean_text'].apply(lambda x: remove_stopwords(x))
ds2.head()
```

REMOVAL OF FREQUENT WORDS

```python
from collections import Counter
word_count=Counter()
for text in ds2['clean_text']:
    for word in text.split():
        word_count[word]+=1

word_count.most_common(10)
```

```python
FREQUENT_WORDS=set(word for (word,wc) in word_count.most_common(5))
def remove_freq_word(text):
    return " ".join([word for word in text.split() if word not in FREQUENT_WORDS])
```

```python
ds2['clean_text']=ds2['clean_text'].apply(lambda x: remove_freq_word(x))
ds2.head()
```

STEMMING

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
ds2['stemmed_text']=ds2['clean_text'].apply(lambda x: stem_words(x))
ds2.head()
```

LEMMETIZATION & POS TAGGING (PROBLEMS)

```python
import nltk
# this is for lemmatization with POS
nltk.download('wordnet')
nltk.download('omw-1.4')
```

LEMMATIZATION (CORRECT)

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatize_word(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
ds2['lematized_text']=ds2['clean_text'].apply(lambda x: lemmatize_word(x))
ds2.head()
```

```python
ds3=ds2.drop(columns=['clean_text','stemmed_text','Message'])
ds3.head()
```

```python
ds3['email']=ds3['lematized_text']
ds4=ds3.drop('lematized_text',axis=1)
ds4.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(ds4.email,ds4.spam,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec=CountVectorizer()
X_train_count=vec.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

```python
# Multinomial Naive Bayes
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(random_state=42)
model1.fit(X_train_count, y_train)
```

```python
# SVM 
from sklearn.svm import SVC
model2 = SVC(kernel='linear', random_state=42)
model2.fit(X_train_count, y_train)
```

```python
X_test_count=vec.transform(X_test) # Accuracy for Multinomial Naive Bayes
model.score(X_test_count,y_test)
```

```python
X_test_count=vec.transform(X_test) # Accuracy for Random Forest
model1.score(X_test_count,y_test)
```

```python
X_test_count=vec.transform(X_test) # Accuracy for SVM
model2.score(X_test_count,y_test)
```

```python
emails=["upto 30% discount of our product.Dont't miss the reward"]
emails_count=vec.transform(emails)
model.predict(emails_count)
```
